```python
import pandas as pd
import numpy as np
import random
```

```python
df=pd.read_csv("_data_2.csv")
```

```python
df
```

```python
df.dtypes
```

```python
for i in df.columns:
    print(i," ", df[i].unique())
```

```python
df.isnull().sum()
```

```python
df.replace(['?','N/A','Unknown','NaN',''],np.nan, inplace=True)
```

```python
df
```

```python
for i in df.columns:
    print(i," ", df[i].unique())
```

```python
df.drop(columns='Garbage',inplace=True)
```

```python
df
```

```python
df['Age'].replace('Twenty-five',25,inplace=True)
```

```python
df
```

```python
for i in df.columns:
    print(i," ", df[i].unique())
```

```python
df['Age']=pd.to_numeric(df['Age'],errors='coerce')
```

```python
df.dtypes
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df['Salary']=pd.to_numeric(df['Salary'], errors='coerce')
```

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df['Salary']=df['Salary'].astype(float)
```

```python
df.dtypes
```

```python
df.dropna(subset=['Name','Gender'],inplace=True)
```

```python
df
```

```python
df.loc[df['Salary']>100000, 'Salary'] = np.nan
```

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
df['Address'] =df['Address'].str.title()
```

```python
df['Address'].unique()
```

```python
df['Gender']=df['Gender'].str.upper()
```

```python
df['Country'].fillna('Unknown', inplace=True)
```

```python
df['Phone Number'] =df['Phone Number'].str.replace('[^0-9]','',regex=True)
df['Phone Number'] = df['Phone Number'].apply(lambda x: f'{x[:3]}-{x[3:6]}-{x[6:]}' if pd.notnull(x) and len(x)==10 else x )
```

```python
df
```

```python
df = df[df['Phone Number'].str.match(r'^\d{3}-\d{3}-\d{4}$',na=False)]
df
```

```python
df['Position'] =df['Position'].str.title()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.drop_duplicates(subset=['Phone Number'], keep='first', inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
age_upper_limit=df['Age'].quantile(0.95)
df.loc[df['Age']>age_upper_limit, 'Age'] = age_upper_limit
df
```

```python
df['Full Address']=df['Address']+','+df['Country']
```

```python
df = df[['Name','Age','Gender','Salary','Address','Country','Position','Phone Number','Full Address']]
```
