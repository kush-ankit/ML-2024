```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
q1=tips['total_bill'].quantile(0.25)
q3=tips['total_bill'].quantile(0.75)
iqr=q3-q1
iqr
```

```python
lower_bound = q1-1.5*iqr
upper_bound = q3+1.5*iqr
print(lower_bound)
print(upper_bound)
```

```python
filtered_tips = tips[(tips['total_bill']>=lower_bound) & (tips['total_bill']<=upper_bound)]
filtered_tips
```

```python
sns.boxplot(x='day',y='total_bill',data=filtered_tips)
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='day',palette='coolwarm')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
data=tips[['total_bill','tip','size']]
corr=data.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```
