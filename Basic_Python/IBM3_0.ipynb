````python
list1 = [1,2,3,7,6,4,0,9,2]
list2 = [9,23,7,83,47,8,9,3,2,4,6,8]

def bubble_sort(list1):
  for i in range(len(list1)-1):
    for j in range(len(list1)-i-1):
      if list1[j] > list1[j+1]:
        list1[j],list1[j+1] = list1[j+1],list1[j]


def merge_list(list1,list2):
  res = []
  for i in list2:
    if i%2 == 0:
      res.append(i)
  list1.extend(res)

merge_list(list1,list2)
bubble_sort(list1)
print(list1)
````

````python
# prompt: ALL THE USECASES OF NUMPY WITH CODE

import numpy as np

# Creating arrays
# --------------------------------------------------------------------------------

# Create an array from a list
arr1 = np.array([1, 2, 3, 4])
print("Array from a list:", arr1)

# Create an array of zeros
arr2 = np.zeros(5)
print("\nArray of zeros:", arr2)

# Create an array of ones
arr3 = np.ones(3)
print("\nArray of ones:", arr3)

# Create an array with a specific range
arr4 = np.arange(10)
print("\nArray with range:", arr4)

# Create an array with evenly spaced values
arr5 = np.linspace(0, 1, 5)
print("\nArray with evenly spaced values:", arr5)

# Create a multi-dimensional array
arr6 = np.array([[1, 2], [3, 4]])
print("\nMulti-dimensional array:\n", arr6)

# Create an array with random values
arr7 = np.random.rand(3, 2)
print("\nArray with random values:\n", arr7)


# Array attributes and manipulation
# --------------------------------------------------------------------------------

# Get the shape of an array
print("\nShape of arr1:", arr1.shape)

# Get the number of dimensions
print("\nNumber of dimensions in arr6:", arr6.ndim)

# Get the data type of an array
print("\nData type of arr1:", arr1.dtype)

# Reshape an array
arr8 = arr4.reshape(2, 5)
print("\nReshaped array:\n", arr8)

# Flatten an array
arr9 = arr6.flatten()
print("\nFlattened array:", arr9)

# Transpose an array
arr10 = np.transpose(arr6)
print("\nTransposed array:\n", arr10)


# Array operations
# --------------------------------------------------------------------------------

# Element-wise operations
arr11 = np.array([5, 6, 7, 8])
print("\nElement-wise addition:", arr1 + arr11)
print("\nElement-wise multiplication:", arr1 * arr11)

# Matrix multiplication
arr12 = np.array([[1, 2], [3, 4]])
arr13 = np.array([[5, 6], [7, 8]])
print("\nMatrix multiplication:\n", np.dot(arr12, arr13))

# Sum of array elements
print("\nSum of arr1:", np.sum(arr1))

# Maximum and minimum values
print("\nMaximum value in arr1:", np.max(arr1))
print("\nMinimum value in arr1:", np.min(arr1))

# Index of maximum and minimum values
print("\nIndex of maximum value in arr1:", np.argmax(arr1))
print("\nIndex of minimum value in arr1:", np.argmin(arr1))

# Unique elements
arr14 = np.array([1, 2, 2, 3, 3, 3, 4])
print("\nUnique elements in arr14:", np.unique(arr14))

# Mathematical functions
print("\nSquare root of arr1:", np.sqrt(arr1))
print("\nLogarithm of arr1:", np.log(arr1))


# Statistical functions
# --------------------------------------------------------------------------------

# Mean
print("\nMean of arr1:", np.mean(arr1))

# Median
print("\nMedian of arr1:", np.median(arr1))

# Standard deviation
print("\nStandard deviation of arr1:", np.std(arr1))

# Variance
print("\nVariance of arr1:", np.var(arr1))

#CUMMULATIVE SUM
print("\nCumulative sum of arr1:", np.cumsum(arr1))

#cummalative product
print("\nCumulative product of arr1:", np.cumprod(arr1))
````

````python
# prompt: all random functions in numpy

# Generating random numbers from various distributions
# --------------------------------------------------------------------------------

# Uniform distribution
uniform_random = np.random.rand(3, 2)  # Values between 0 and 1
print("\nUniform distribution:\n", uniform_random)

# Normal distribution
normal_random = np.random.randn(4, 3)  # Mean 0, standard deviation 1
print("\nNormal distribution:\n", normal_random)

# Integer random numbers
integer_random = np.random.randint(1, 10, size=(2, 5))  # Integers between 1 and 9
print("\nInteger random numbers:\n", integer_random)

# Choice from a given array
choices = np.random.choice([2, 4, 6, 8], size=10)
print("\nRandom choices from an array:", choices)

# Shuffle an array in-place
arr_to_shuffle = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr_to_shuffle)
print("\nShuffled array:", arr_to_shuffle)

# Generate random permutation of an array
permuted_array = np.random.permutation([1, 2, 3, 4, 5])
print("\nPermuted array:", permuted_array)

# Set random seed for reproducibility
np.random.seed(42)
random_with_seed = np.random.rand(2, 3)
print("\nRandom numbers with seed:\n", random_with_seed)
````

````python
import pandas as pd

df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value1': [1, 2, 3, 4]})
df2 = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value2': [5, 6, 7, 8]})

# Left Join
left_join = pd.merge(df1, df2, on='key', how='left')
print("Left Join:\n", left_join)

# Inner Join
inner_join = pd.merge(df1, df2, on='key', how='inner')
print("\nInner Join:\n", inner_join)

# Outer Join (Full Outer Join)
outer_join = pd.merge(df1, df2, on='key', how='outer')
print("\nOuter Join:\n", outer_join)
````
